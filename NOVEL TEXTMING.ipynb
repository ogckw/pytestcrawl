```python
#ck101 text mining
import requests
from bs4 import BeautifulSoup as bs
import jieba
headers  = {'Host':'ck101.com',
'Referer':'http://ck101.com/thread-2776008-211-1.html',
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/45.0.2454.93 Safari/537.36'}

res = requests.get('http://ck101.com/thread-2776008-210-1.html', headers = headers)
soup = bs(res.text)
print soup.select('.t_f')
dic = {}




for ent in  soup.select('.t_f'):
    for w in jieba.cut(ent.select('.title')[0].text.strip()):
        if w not in dic:
            dic[w] = 1
        else:
            dic[w] = dic[w] + 1
            
for ele in  dic:   
    print ele, dic[ele] 
```
